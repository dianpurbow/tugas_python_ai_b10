**0. Import module yang diperlukan**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style untuk visualisasi seaborn
sns.set_theme(style="whitegrid")
import warnings
warnings.filterwarnings('ignore')# Write your code here
```

**1. Membaca data**


> Silakan gunakan teknik membaca data teratas dan terbawah (cth: head(), tail()) yang tersedia pada pandas

> Cek tipe data dan missing value

> Tampilkan statistik deskriptif

```python
# Membaca dataset
# Pastikan file 'worldometer_data.csv' sudah di-upload ke direktori yang sama
df = pd.read_csv('worldometer_data.csv')

# Menampilkan 5 data teratas dan terbawah
display(df.head())
display(df.tail())
```

```python
# Mengecek tipe data dan total missing values per kolom
print("Info Dataset:")
display(df.info())

print("\nMissing Values per Kolom:")
display(df.isnull().sum())
```

```python
# Menampilkan statistik deskriptif dari kolom numerik
display(df.describe())
```

Insight :

> Dataset ini memiliki rentang nilai yang sangat besar (terlihat dari nilai max yang sangat jauh dari nilai rerata/mean pada kolom TotalCases), menandakan adanya skewness (kemiringan) yang tajam.

Terdapat beberapa missing values (NaN) pada kolom seperti NewCases, TotalRecovered, dan ActiveCases. Hal ini wajar karena tidak semua negara melaporkan pembaruan data secara real-time atau memiliki rekam data historis yang lengkap.

Tipe data sebagian besar sudah dalam bentuk numerik (float64/int64), kecuali kolom seperti Country/Region dan Continent yang bertipe object (kategorikal).


**2. Analisis Distribusi**

> Plot distribusi TotalCases, TotalDeaths, dan TotalRecovered

> Buat histogram log-skala untuk kolom TotalCases

> Bandingkan populasi terhadap jumlah kasus (scatter plot)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df['TotalCases'].dropna(), bins=30, ax=axes[0], color='blue', kde=True)
axes[0].set_title('Distribusi Total Cases')

sns.histplot(df['TotalDeaths'].dropna(), bins=30, ax=axes[1], color='red', kde=True)
axes[1].set_title('Distribusi Total Deaths')

sns.histplot(df['TotalRecovered'].dropna(), bins=30, ax=axes[2], color='green', kde=True)
axes[2].set_title('Distribusi Total Recovered')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalCases'].dropna(), bins=30, log_scale=True, kde=True, color='purple')
plt.title('Log-Scale Histogram dari Total Cases')
plt.xlabel('Total Cases (Log Scale)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Population', y='TotalCases', alpha=0.6, edgecolor=None)
plt.xscale('log')
plt.yscale('log')
plt.title('Populasi vs Total Kasus (Log Scale)')
plt.xlabel('Populasi (Log Scale)')
plt.ylabel('Total Kasus (Log Scale)')
plt.show()
```

Insight :

> Distribusi asli dari TotalCases, Deaths, dan Recovered sangat right-skewed (condong ke kanan), membuktikan bahwa mayoritas negara memiliki jumlah kasus yang relatif rendah, sementara segelintir negara memonopoli angka kasus tertinggi secara global.

Penggunaan transformasi log-skala membuat kurva distribusi menjadi lebih berdistribusi normal, sehingga variasi data lebih mudah diamati.

Dari scatter plot, terlihat adanya korelasi positif antara populasi dan total kasus. Namun, karena titik-titik data agak menyebar luas, ada faktor lain (seperti kebijakan lockdown atau jumlah pengujian) yang memengaruhi jumlah kasus, bukan murni hanya dari besarnya populasi.


**3. Analisis Per Benua (Continent)**

> Hitung total kasus, kematian, dan kesembuhan per benua

> Visualisasikan perbandingan kasus dan kematian antar benua

```python
continent_stats = df.groupby('Continent')[['TotalCases', 'TotalDeaths', 'TotalRecovered']].sum().reset_index()
display(continent_stats.sort_values(by='TotalCases', ascending=False))
```

```python
continent_melt = continent_stats.melt(id_vars='Continent',
                                      value_vars=['TotalCases', 'TotalDeaths'],
                                      var_name='Metric', value_name='Total')

plt.figure(figsize=(12, 6))
sns.barplot(data=continent_melt, x='Continent', y='Total', hue='Metric', palette='Set2')
plt.title('Perbandingan Total Cases dan Total Deaths per Benua')
plt.ylabel('Jumlah (Log Scale)')
plt.yscale('log') # Menggunakan log scale agar bar kematian terlihat jelas bersanding dengan kasus
plt.show()
```

Insight :

>Berdasarkan agregasi, benua Eropa dan Amerika Utara mendominasi sebagai episentrum dengan total kasus dan kematian kumulatif terbanyak.

Secara visual (menggunakan log scale), gap antara kematian dan jumlah kasus cukup konsisten antar benua. Australia/Oceania mencatatkan rekor metrik terendah secara global untuk penyebaran virus ini.


**4. Analisis Tes dan Positivity Rate**
> Buat kolom baru "Positivity Rate" = TotalCases / TotalTests

> Analisis negara-negara dengan positivity rate tertinggi

```python
# Mencegah division by zero
df['Positivity Rate'] = df['TotalCases'] / df['TotalTests']
```

```python
top_positivity = df[['Country/Region', 'Positivity Rate', 'TotalCases', 'TotalTests']].dropna()
top_positivity = top_positivity.sort_values(by='Positivity Rate', ascending=False).head(10)
display(top_positivity)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_positivity, x='Positivity Rate', y='Country/Region', palette='Reds_r')
plt.title('Top 10 Negara dengan Positivity Rate Tertinggi')
plt.show()
```

Insight :

> Positivity Rate adalah rasio kasus positif berbanding total jumlah tes. Angka yang mendekati atau bahkan melebihi rasio rasional seringkali mengindikasikan under-testing (negara hanya mengetes orang yang sudah masuk rumah sakit/bergejala parah).

Negara-negara yang muncul dalam visualisasi ini seringkali adalah negara berkembang atau negara dengan infrastruktur medis yang kewalahan merespons lonjakan.


**5. Death Rate & Recovery Rate**
> Buat kolom baru "Death Rate" = TotalDeaths / TotalCases

> Buat kolom "Recovery Rate" = TotalRecovered / TotalCases

> Bandingkan death rate dan recovery rate secara global

```python
df['Death Rate'] = df['TotalDeaths'] / df['TotalCases']
df['Recovery Rate'] = df['TotalRecovered'] / df['TotalCases']
```

```python
global_death_rate = df['TotalDeaths'].sum() / df['TotalCases'].sum()
global_recovery_rate = df['TotalRecovered'].sum() / df['TotalCases'].sum()

print(f"Global Death Rate: {global_death_rate:.2%}")
print(f"Global Recovery Rate: {global_recovery_rate:.2%}")
```

```python
rates = pd.DataFrame({
    'Metric': ['Global Death Rate', 'Global Recovery Rate'],
    'Rate': [global_death_rate, global_recovery_rate]
})

plt.figure(figsize=(6, 5))
sns.barplot(data=rates, x='Metric', y='Rate', palette='viridis')
plt.title('Perbandingan Global Death Rate dan Recovery Rate')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
plt.show()
```

Insight :

> Secara global, virus ini memiliki tingkat kesembuhan (Recovery Rate) yang jauh lebih tinggi secara signifikan dibandingkan tingkat kefatalannya (Death Rate).

Hal ini membuktikan bahwa meskipun penyebarannya sangat masif, peluang kesintasannya juga tinggi.


**6. Serious/Critical Case Analysis**

> Visualisasi negara dengan jumlah kasus kritis tertinggi

> Bandingkan dengan jumlah kasus aktif

```python
top_critical = df[['Country/Region', 'Serious,Critical', 'ActiveCases']].dropna()
top_critical = top_critical.sort_values(by='Serious,Critical', ascending=False).head(10)
```

```python
critical_melt = top_critical.melt(id_vars='Country/Region',
                                  value_vars=['ActiveCases', 'Serious,Critical'],
                                  var_name='Kategori', value_name='Jumlah')

plt.figure(figsize=(12, 6))
sns.barplot(data=critical_melt, x='Country/Region', y='Jumlah', hue='Kategori', palette='magma')
plt.title('Top 10 Negara Kasus Kritis vs Kasus Aktif')
plt.xticks(rotation=45)
plt.show()
```

Insight :

> Dari visualisasi ini, terlihat bahwa persentase kasus kritis (Serious/Critical) rata-rata adalah fraksi kecil dari keseluruhan kasus aktif (Active Cases).

Amerika Serikat (USA) memiliki bar paling tinggi di antara negara lain, menggambarkan besarnya beban sistem gawat darurat (ICU) di rumah sakit negara tersebut pada momen data ditarik.

**8. Correlation Matrix**

> Buat heatmap korelasi untuk kolom numerik

> Apa yang menarik dari korelasi antara populasi, test, dan kasus?

```python
# Memfilter hanya kolom numerik agar tidak error saat korelasi
numerical_cols = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_cols.corr()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Fitur Numerik COVID-19 Dataset', fontsize=15)
plt.show()
```

Insight :

> Terdapat korelasi yang nyaris sempurna (r ~ 0.99) antara TotalCases, TotalDeaths, dan TotalRecovered. Semakin tinggi kasus infeksi di suatu negara, maka angka kematian dan kesembuhan secara absolut akan ikut terkerek naik.

Korelasi antara Population dengan parameter penyebaran kasus bernilai sedang-kuat (r ~ 0.5 - 0.6). Ini menguatkan analisis scatter plot di awal: meskipun populasi besar rentan memiliki lebih banyak kasus, keberhasilan mitigasi dan kapasitas testing juga memainkan peran krusial di luar murni soal populasi.
